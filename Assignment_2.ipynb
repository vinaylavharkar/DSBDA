```python
Q1 = data['First year:   Sem 1'].quantile(0.25)
Q3 = data['First year:   Sem 1'].quantile(0.75)
IQR = Q3 - Q1
data = data[(data['First year:   Sem 1'] >= Q1 - 1.5 * IQR) & (data['First year:   Sem 1'] <= Q3 + 1.5 * IQR)]

print(data['First year:   Sem 1'].head())
```
