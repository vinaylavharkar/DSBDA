```python
data.isnull().sum()
```

```python
data.replace(np.nan, data['age'].mean, inplace=True)
data.replace(np.nan, data['embark_town'].mean, inplace=True)
data.replace(np.nan, data['embarked'].mean, inplace=True)
```

```python
data['deck'].fillna(data['deck'].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
data['survived'].count()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['fare'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Ticket Prices')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
